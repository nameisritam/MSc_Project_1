```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel(r'C:\Users\ratha\cleaned_data_2.xlsx')

# Create a new DataFrame with specific columns
# Ensure the column names match exactly with your Excel file
data = df[['year', 'Currency in circulation -Total ','Reserve Money (Liabilities/Components) ','Net foreign exchange assets of RBI ']]

# Convert the 'year' column to datetime format (if it's in a year format)
data['year'] = pd.to_datetime(data['year'], format='%Y', errors='coerce')

# Plot the data for 'Currency in circulation -Total'
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['Currency in circulation -Total '], marker=',', linestyle='-', color='r')

# Customize the plot
plt.title('Currency in Circulation Over Time')
plt.xlabel('Year')
plt.ylabel('Currency in Circulation')
plt.grid(True)

# Show the plot
plt.show()

# Plot the data

plt.plot(data['year'], data['Reserve Money (Liabilities/Components) '], marker=',', linestyle='-', color='r')

# Customize the plot
plt.title('Reserve Money (Liabilities/Components) over time' )
plt.xlabel('Year')
plt.ylabel('Reserve Money (Liabilities/Components)')
plt.grid(True)

# Show the plot
plt.show()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['Net foreign exchange assets of RBI '], marker=',', linestyle='-', color='r')

# Customize the plot
plt.title('Net foreign exchange assets of RBI over time' )
plt.xlabel('Year')
plt.ylabel('Net foreign exchange assets of RBI ')
plt.grid(True)

# Show the plot
plt.show()
```
